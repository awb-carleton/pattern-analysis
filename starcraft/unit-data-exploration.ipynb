```python
import sc2reader.data as sc2data
from itertools import groupby
import os
replay = pvt
zerg_fields = ['baneling', 'banelingnest', 'broodlord', 'corruptor', 'drone', 'evolutionchamber', 'extractor', 'greaterspire', 'hatchery', 'hive', 'hydralisk', 'hydraliskden', 'infestationpit', 'infestedterran', 'infestor', 'lair', 'locust', 'lurker', 'lurkerden', 'mutalisk', 'nydusnetwork', 'nydusworm', 'overlord', 'overseer', 'queen', 'ravager', 'roach', 'roachwarren', 'spawningpool', 'spinecrawler', 'spire', 'sporecrawler', 'swarmhost', 'ultralisk', 'ultraliskcavern', 'viper', 'zergling']
protoss_fields = ['adept', 'archon', 'assimilator', 'carrier', 'colossus', 'cyberneticscore', 'darkshrine', 'darktemplar', 'disruptor', 'fleetbeacon', 'forge', 'gateway', 'hightemplar', 'immortal', 'interceptor', 'mothership', 'mothershipcore', 'nexus', 'observer', 'oracle', 'phoenix', 'photoncannon', 'probe', 'pylon', 'reactor', 'roboticsbay', 'roboticsfacility', 'sentry', 'stalker', 'stargate', 'tempest', 'templararchive', 'twilightcouncil', 'voidray', 'warpgate', 'warpprism', 'zealot']
terran_fields = ['armory', 'banshee', 'barracks', 'barrackstechlab', 'barracksreactor', 'battlecruiser', 'battlehellion', 'bunker', 'commandcenter', 'cyclone', 'engineeringbay', 'factory', 'factoryreactor', 'factorytechlab', 'fusioncore', 'ghost', 'ghostacademy', 'hellion', 'marauder', 'marine', 'medivac', 'missileturret', 'mule', 'orbitalcommand', 'planetaryfortress', 'raven', 'reaper', 'refinery', 'scv', 'sensortower', 'siegetank', 'starport', 'starportreactor', 'starporttechlab', 'supplydepot', 'techlab', 'thor', 'viking', 'warhound', 'widowmine']
fields = zerg_fields + protoss_fields + terran_fields
series = np.zeros((replay.frames, len(protoss_fields)))
# [[t.name for t in e.unit.type_history.values()] for e in pvt.events + zvz.events if isinstance(e, sc2reader.events.UnitDoneEvent) or isinstance(e, sc2reader.events.UnitBornEvent) if (e.unit.is_army or e.unit.is_building or e.unit.is_worker) and list(e.unit.type_history.values())[0].name.lower() not in fields]
frame_to_events = {frame: list(events) for frame, events in groupby(sorted(replay.events, key=lambda e: e.frame), lambda e: e.frame)}
for frame in range(len(series)):
    if frame > 0:
        series[frame] = series[frame - 1]
    for e in frame_to_events.get(frame, []):
        if not hasattr(e, "unit") or e.unit.owner != replay.players[0]:
            continue
        if isinstance(e, sc2reader.events.UnitBornEvent) or isinstance(e, sc2reader.events.UnitDoneEvent):
            unit = e.unit
            if unit.is_army or unit.is_worker or unit.is_building:
                name = sorted(unit.type_history.items())[0][1].name.lower()
                assert name in protoss_fields
                series[frame][protoss_fields.index(name)] += 1
        if isinstance(e, sc2reader.events.UnitDiedEvent):
            unit = e.unit
            assert unit.is_army or unit.is_worker or unit.is_building
            name = sorted(unit.type_history.items())[0][1].name.lower()
            assert name in protoss_fields
            series[frame][protoss_fields.index(name)] -= 1
fig, ax = plt.subplots(figsize=(20,10))
colors = plt.cm.get_cmap("tab20").colors + plt.cm.get_cmap("tab20b").colors
for j in range(series.shape[1]):
    ax.plot(np.arange(replay.frames), series[:, j], color=colors[j], alpha=0.7)
# line_legend = ax.legend(protoss_fields, loc="upper center", ncol=7, fancybox=True, shadow=True,
#                         bbox_to_anchor=(0.3, 1.05))
fig.tight_layout()
plt.draw()
cols = {j for j in np.nonzero(series)[1]}
for j in cols:
    field = protoss_fields[j]
    if os.path.exists(f"sprites/{field}.png"):
        im = plt.imread(f"sprites/{field}.png")
        peak = series[:, j].argmax()
        xo, yo = ax.transData.transform((peak, series[peak, protoss_fields.index(field)]))
#         print((peak, series[peak, protoss_fields.index(field)]), xo, yo, ax.get_xlim(), ax.get_ylim(), fig.canvas.get_width_height())
        fig.figimage(im, xo=xo - im.shape[0] // 2, yo=yo - im.shape[1] // 2)
        plt.draw()
```
